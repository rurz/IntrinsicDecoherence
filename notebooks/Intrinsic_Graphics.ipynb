```julia
    α = 4
	ω = 4
	λ = 0.5
	
	ωₘ = ω - λ
	ωₚ = ω + λ
	
	Ω(g) = (1/2) * (ω + ωₚ - √(8 * g^2 + (ωₚ - ω)^2))
	Ω₂(g) = (1/2) * (ω + ωₚ + √(8 * g^2 + (ωₚ - ω)^2))
	
	c2ϕ(g) = (ωₚ - ω)/√(8 * g^2 + (ωₚ - ω)^2)
	s2ϕ(g) = (2 * √2 * g)/√(8 * g^2 + (ωₚ - ω)^2)
	
	cc2ϕ(g) = (1/2) * (1 + c2ϕ(g))
	ss2ϕ(g) = (1/2) * (1 - c2ϕ(g))
;
```

```julia
expΩΩ₂(γ, g,t) = 
exp(
    big(γ * t * 
        exp(
            (1im/γ) * (
                Ω(g) - Ω₂(g)
                )
            )
        )
)
```

```julia
exv_a(g, γ, t) = 
abs2(α) * cc2ϕ(g) * ss2ϕ(g) *

	(1 - (1/2) * exp(-γ * t) * (
        
        expΩΩ₂(γ, g, t) + conj(expΩΩ₂(γ, g, t))
        
        )
    
    )
```

```julia
t_list = range(0, stop = 100, length = 1000)
```

```julia
texv_a = [real(exv_a(0.1, 10, t_list[τ])) for τ in 1:length(t_list)]
```
