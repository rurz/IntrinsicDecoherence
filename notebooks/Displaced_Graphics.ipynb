# 1. Coherent states average

```julia
const α = 4
const ω = 4;
```

```julia
t1 = range(0, stop = 7, length = 500);
```

## 1.1 Position quadrature operator

```julia
tann1 = [quad_a(α, 0.1, ω, 10, t1[τ]) for τ in 1:length(t1)];
tann2 = [quad_a(α, 0.7, ω, 10, t1[τ]) for τ in 1:length(t1)];
tann3 = [quad_a(α, 1.5, ω, 10, t1[τ]) for τ in 1:length(t1)];
```

```julia
	figreim1 = figure(figsize = (10, 6))
	axreim1 = gca()
	
	axreim1.plot(t1, real(tann1), "r", lw = 3, label = L"\lambda = 0.1")
	axreim1.plot(t1, real(tann2), "g", lw = 3, label = L"\lambda = 0.7")
	axreim1.plot(t1, real(tann3), "b", lw = 3, label = L"\lambda = 1.5")
	
	#axreim1.set_ylim([-9, 9])
	
	axreim1.legend(fontsize = 18)
	axreim1.set_ylabel(L"\rm{Amplitude}", fontsize = 20)
	axreim1.set_xlabel(L"t\; [s]", fontsize = 18)
	setp(axreim1.get_xticklabels(), fontsize = 18)
	setp(axreim1.get_yticklabels(), fontsize = 18)
	axreim1.set_title(L"Expectation value of $\langle\hat{a}^{\dagger}+\hat{a}\rangle$", fontsize = 18)
	
	tight_layout()
	
	#savefig("quadex.pdf", dpi = 600)
```

## 1.2 Number operator

```julia
tnum1 = [num_a(α, 0.1, ω, 10, t1[τ]) for τ in 1:length(t1)];
tnum2 = [num_a(α, 0.7, ω, 10, t1[τ]) for τ in 1:length(t1)];
tnum3 = [num_a(α, 1.5, ω, 10, t1[τ]) for τ in 1:length(t1)];
```

```julia
	fignum = figure(figsize = (10, 6))
	axnum = gca()
	
	axnum.plot(t1, real(tnum1), "r", lw = 3, label = L"\lambda = 0.1")
	axnum.plot(t1, real(tnum2), "g", lw = 3, label = L"\lambda = 0.7")
	axnum.plot(t1, real(tnum3), "b", lw = 3, label = L"\lambda = 1.5")
	
	axnum.set_ylim([16, 21.5])
	
	axnum.legend(fontsize = 18)
	axnum.set_ylabel(L"\rm{Number of Photons}", fontsize = 20)
	axnum.set_xlabel(L"t\; [s]", fontsize = 18)
	setp(axnum.get_xticklabels(), fontsize = 18)
	setp(axnum.get_yticklabels(), fontsize = 18)
	axnum.set_title(L"Expectation value of $\langle\hat{a}^{\dagger}\hat{a}\rangle$", fontsize = 20)
	
	tight_layout()
	
	#savefig("nex.pdf", dpi = 600)
```

# 2. Squeezed states average

```julia
const αₛ = 4
const r = 0.3
const ωₛ = 4;
```

```julia
tlist = range(0, stop = 7, length = 500);
```

## 2.1 Position quadrature operator

```julia
tsq1 = [quads_a(αₛ, r, 0.7, 0.7, ωₛ, 10, tlist[τ]) for τ in 1:length(tlist)];
tsq2 = [quads_a(αₛ, r, 1.4, 0.7, ωₛ, 10, tlist[τ]) for τ in 1:length(tlist)];
tsq3 = [quads_a(αₛ, r, 2.1, 0.7, ωₛ, 10, tlist[τ]) for τ in 1:length(tlist)];
```

```julia
	figreim1 = figure(figsize = (10, 6))
	axreim1 = gca()
	
	#axreim1.plot(tlist, real(ctrs), "y--", lw = 2, alpha = 0.5)
	axreim1.plot(tlist, real(tsq1), "r", lw = 3, label = L"\theta = 0.7")
	axreim1.plot(tlist, real(tsq2), "g", lw = 3, label = L"\theta = 1.4")
	axreim1.plot(tlist, real(tsq3), "b", lw = 3, label = L"\theta = 2.1")
	
	#axreim1.set_ylim([-9, 9])
	
	#axreim1.text(1,4, L"r = 0.3", fontsize = 20)
	
	axreim1.legend(fontsize = 18)
	axreim1.set_ylabel(L"\rm{Amplitude}", fontsize = 20)
	axreim1.set_xlabel(L"t\; [s]", fontsize = 18)
	setp(axreim1.get_xticklabels(), fontsize = 18)
	setp(axreim1.get_yticklabels(), fontsize = 18)
	axreim1.set_title(L"Expectation value of $\langle\hat{a}^{\dagger}+\hat{a}\rangle$", fontsize = 18)
	
	tight_layout()
	
	#savefig("quads.pdf", dpi = 600)
```

## 2.2 Number operator

```julia
ctrS = [nums_a(α, 0, 0, 0.7, ω, 10, tlist[τ]) for τ in 1:length(tlist)];
tSq1 = [nums_a(α, r, 0.7, 0.7, ω, 10, tlist[τ]) for τ in 1:length(tlist)];
tSq2 = [nums_a(α, r, 1.4, 0.7, ω, 10, tlist[τ]) for τ in 1:length(tlist)];
tSq3 = [nums_a(α, r, 2.1, 0.7, ω, 10, tlist[τ]) for τ in 1:length(tlist)];
```

```julia
	figreim2 = figure(figsize = (10, 6))
	axreim2 = gca()
	
	axreim2.plot(tlist, real(ctrS), "y--", lw = 2, alpha = 0.5)
	axreim2.plot(tlist, real(tSq1), "r", lw = 3, label = L"\theta = 0.7")
	axreim2.plot(tlist, real(tSq2), "g", lw = 3, label = L"\theta = 1.4")
	axreim2.plot(tlist, real(tSq3), "b", lw = 3, label = L"\theta = 2.1")
	
	#axreim1.set_ylim([-9, 9])
	
	#axreim2.text(5, 17.2, L"r = \theta = 0", fontsize = 16)
	axreim2.annotate(L"r = \theta = 0", xy=(5, 17.5), xytext=(5.5, 16), arrowprops = Dict("arrowstyle"=>"->"), fontsize = 14)
	
	axreim2.legend(fontsize = 18, loc = 1)
	axreim2.set_ylabel(L"\rm{Number of Photons}", fontsize = 20)
	axreim2.set_xlabel(L"t\; [s]", fontsize = 18)
	setp(axreim2.get_xticklabels(), fontsize = 18)
	setp(axreim2.get_yticklabels(), fontsize = 18)
	axreim2.set_title(L"Expectation value of $\langle\hat{a}^{\dagger}\hat{a}\rangle$", fontsize = 18)
	
	tight_layout()
	
	#savefig("nums.pdf", dpi = 600)
```
